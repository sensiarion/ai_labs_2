Matplotlib поставляется с настройками по умолчанию, которые позволяют настраивать различные свойства. Вы можете управлять значениями по умолчанию почти каждого свойства в matplotlib: размер фигуры, количество точек на дюйм, ширина линии, цвет и стиль осей, свойства оси и сетки, свойств текста и шрифта и так далее. Значения по умолчанию для matplotlib достаточно хороши в большинстве случаев и изменять некоторые свойства нужно только для конкретных случаев.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

В приведенной ниже ячейке реализованы все параметры фигуры, которые влияют на внешний вид графика. Настройки были явно установлены на значения по умолчанию, но теперь можно в интерактивном режиме играть со значениями, чтобы изучить их влияние.

```python
# импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Создать новую фигуру размером 8x6 точек, используя 100 точек на дюйм
plt.figure(figsize=(8,6), dpi=100)

# Создание нового заголовка
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Построение Косинуса синим цветом с непрерывной линией шириной 1 (пиксели)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# Синус участка с использованием зеленого цвета с непрерывной линией шириной 1 (пикселей)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Установить лимит для X
plt.xlim(-4.0,4.0)

# Установить количество тиков для X
plt.xticks(np.linspace(-4,4,9,endpoint=True))

#Установить лимит для Y
plt.ylim(-1.0,1.0)

# Установить количество тиков для Y
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# Сохранить рисунок с разрешением 72 dpi
# savefig("figure.png",dpi=72)

# Отобразить результат
plt.show()
```

Добавим в верхнем левом углу легенду. Для этого требуется только добавления ключевой метки аргумента (которая будет использоваться в поле легенды) к командам графика.

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="Косинус")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="Синус")

plt.legend(loc='upper left', frameon=False)

# Отобразить результат
plt.show()
```

Таким образом, **Matplotlib** позволяет гибко настраивать отображение графиков.

Еще один бесплатный пакет, основанный на этой библиотеке визуализации данных - это **Seaborn**, которая предоставляет высокоуровневый интерфейс для рисования статистической графики.

Рассмотрим простой пример:

Когда вы работаете с Seaborn, вы можете либо использовать один из встроенных наборов данных, которые имеются в библиотеке, либо можно загрузить Pandas DataFrame. Чтобы начать работать со встроенным набором данных Seaborn, вы можете использовать функцию load_dataset (), как показано в ячейке нижне. Чтобы получить обзор имеющихся наборов данных перейдите по ссылке https://github.com/mwaskom/seaborn-data.

```python
# Импортируем необходимые библиотеки
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем данные
tips = sns.load_dataset("tips")

# Создаем необходимый формат
sns.violinplot(x = "total_bill", data=tips)

# Отображаем
plt.show()
```

В большинстве случаев преимущество визуализации данных заключается в том, что работа будет происходить с собственными данными, а не со встроенными наборами данных библиотеки Seaborn. Seaborn лучше всего работает с Pandas DataFrames и массивами, которые содержат целый набор данных. 

**Помните**, что DataFrames - это способ хранения данных в прямоугольных сетках, которые легко просматривать. Каждая строка этих сеток соответствует измерениям или значениям экземпляра, в то время как каждый столбец представляет собой вектор, содержащий данные для определенной переменной. Это означает, что строки DataFrame не обязательно должны , но могут содержать значения одного и того же типа: они могут быть числовыми, символьными, логическими и т. Д. DataFrames, как уже известно, поставляются с библиотекой Pandas и определяются как двумерные маркированные структуры данных со столбцами потенциально разных типов.

### Тестовый пример
Рассмотрим на основе набора данных из ресторана о оставленных чаевых
Имеются следующие признаки:
* Общий счет (Total bill)
* Чаевые (Tips)
* Пол (Sex)
* Курящий (Smoker)
* День недели (Day)
* Время (Time)
* Количество (Size)

```python
import pandas as pd
data=pd.read_csv('tips.csv')

# Выводим первые 5 записей набора данных
data.head()
```

```python
# Смотрим на количество имеющихся данных
len(data)
```

```python
# Реализуем график, что бы посмотреть на взаимосвязь чаевых от общего счета
sns.regplot(x='tip', y='total_bill', data=data, ci=None)
plt.show()
```

```python
# Строим гистограмму для общего счета в ресторане
plt.hist(data['total_bill'])
plt.show()
```

```python
# Изменим свойства hist
sns.distplot(data['total_bill'], kde=False, rug=True)
plt.show()

# Таким образом получим более наглядую гистограмму
```

```python
# Строим boxplot для Общего счета и дня недели
sns.set_style('darkgrid')
sns.boxplot(x='day', y='total_bill', data=data, palette='Blues')
plt.show()
```

```python
# Строим boxplot для Пола и оставленных чаевых
sns.boxplot(x='sex', y='tip', data=data, palette='Blues')
plt.show()
```

```python
# Добавляем параметры и сохраняем рисунок
fig, ax= plt.subplots()
ax.scatter(data[data['sex']=='Male']['tip'], data[data['sex']=='Male']['total_bill'], color='blue', label='Муж')
ax.scatter(data[data['sex']=='Female']['tip'], data[data['sex']=='Female']['total_bill'], color='green', label='Жен')
ax.set_title('Общий счет и чаевые с распределением по полу', fontdict={'fontsize': 14, 'fontweight': 'bold'})
ax.set_ylabel('Общий счет')
ax.set_xlabel('Чаевые')
ax.legend()
plt.show()

fig.savefig('tips.png', format='png', dpi=250)
```

```python
data=pd.read_csv('premier_league.csv')
df = data.copy()
# Выводим первые 5 записей набора данных
data.head()
```

Для начала просмотрим данные на явные пропуски

```python
df.info()
```

В среднем команды чаще забивают в домашних матчах

Составим таблицу с статистиками игр для каждой команды

```python
home_played = pd.DataFrame(df.groupby(by='home_team').result.count() )
home_played = home_played.rename(columns={'result': 'home played'})

home_won = pd.DataFrame(df[(df['result'] == 'H')].groupby(by='home_team').result.count() )
home_won = home_won.rename(columns={'result': 'home won'})

away_played = pd.DataFrame(df.groupby(by='away_team').result.count() )
away_played = away_played.rename(columns={'result': 'away played'})
away_played.index.rename('home_team',inplace=True)

away_won = pd.DataFrame(df[(df['result'] == 'A')].groupby(by='away_team').result.count() )
away_won = away_won.rename(columns={'result': 'away won'})
away_won.index.rename('home_team',inplace=True)

drafts_count = pd.DataFrame(df[(df['result'] == 'D')].groupby(by='away_team').result.count()) + pd.DataFrame(df[(df['result'] == 'D')].groupby(by='home_team').result.count())
drafts_count.index.rename('home_team',inplace=True)
drafts_count = drafts_count.rename(columns={'result':'drafts'})

games_count = pd.DataFrame(home_played['home played'] + away_played['away played'])
games_count = games_count.rename(columns={0:'played'})


stats = games_count.join(home_played, on='home_team').join(home_won, on='home_team').join(away_played, on='home_team').join(away_won, on='home_team').join(drafts_count, on='home_team')
stats.index.rename('team', inplace=True)

stats = stats.fillna(value=0)
stats['away won'] = stats['away won'].astype('int')
```

Найдём команды с самым большой долей побед

```python
stats['win rate'] = (stats['home won'] + stats['away won']) / stats.played
stats = stats.sort_values(by='win rate', ascending=False)
stats.head(3)
```

Из данной таблицы сложно понять зависимость с количеством игр, добавим информацию о кол-ве сезонов, в которых успела поучаствовать команда

```python
state_with_seasons = stats.join(data.groupby(by='home_team').season.unique())
# state_with_seasons['season_count'] = 0
seasons_count = pd.DataFrame(state_with_seasons.apply(func=lambda x: len(x['season']), axis=1))

state_with_seasons = state_with_seasons.join(seasons_count,on='team')
state_with_seasons = state_with_seasons.drop(columns=['season'])
state_with_seasons = state_with_seasons.rename(columns={0: 'seasons_count'})    
```

Теперь можно посмотреть какие команды играют дольше всего

```python
t = state_with_seasons.sort_values(by='seasons_count', ascending=True)
fig, ax = plt.subplots(figsize=(11,9))
sns.barplot(ax=ax,y=t.index, x=t.seasons_count)
```

Как и ожидалось, самые известные команды отыграли больше всего сезонов

Посмотрим на кол-во выйгранных игр в зависимости от сыгранных игр (визуализация винтерйта)

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.set_ylabel('Побед')
ax.set_xlabel('Игр')
plt.scatter(t.played,t['home won'])
```

Можно увидеть что зависиомость не являтся линейной, а имеет сдвиг у команд с наибольшим числом игр, что указывает на их больший процент побед. В совокупности с предыдущим графиком можно сделать предположение, что хорошая результативность и позволяет им оставаться на плаву большое кол-во времени.

Чтобы быть более уверенным в выдвинутой гипотезе можно рассмотреть % побед команд для каждого сезона отдельно

```python
seasons = df.season.unique()

cols = ['team','played','wins','season']
year_data = pd.DataFrame(columns=cols)

for i in seasons:
    for name in df.home_team.unique():
        try:
            played = pd.DataFrame(df[ (df.season==i) & (df.home_team==name) ].groupby(by='home_team').result.count() )
            w_h = pd.DataFrame(df[ (df.season==i) & (df.home_team==name) & (df.result == 'H')].groupby(by='home_team').result.count() )
            w_a = pd.DataFrame(df[ (df.season==i) & (df.away_team==name) & (df.result == 'A')].groupby(by='away_team').result.count())
            year_data = year_data.append({'team':played.index[0],
                'played':played.result[0],
                'wins':w_h.result[0]+w_a.result[0],
                'season':i}, ignore_index=True)
        except Exception as e:
            pass
```

Рассмотрим на примере 2 долгоиграющих, 2 средних и 2 мало отыгравших команд

```python
low = state_with_seasons[state_with_seasons.seasons_count == 12].head(2)
med = state_with_seasons[(state_with_seasons.seasons_count >= 5) & (state_with_seasons.seasons_count <= 6)].head(2)
long = state_with_seasons[state_with_seasons.seasons_count < 3].head(2)
temp = low.append(med).append(long)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.set_ylabel('% побед')
ax.set_xlabel('сезон')

for index, i in temp.iterrows():
    cur = year_data[year_data.team==i.name]
    plt.plot(cur.season,cur.wins/(cur.played*2))
labels = list(map(lambda x: x[0]+' '+str(x[1]),zip(temp.index,temp.seasons_count)))
ax.legend(labels)
```

На данном графике частично можно увидеть подтверждение гипотезы, но тут слишком мало команд, а вывести все команды одновременно довольно проблематично. Давайте рассмотрим каждую категорию отдельно

```python
temp = state_with_seasons[state_with_seasons.seasons_count == 12]
fig, ax = plt.subplots(figsize=(12,6))
ax.set_ylabel('% побед')
ax.set_xlabel('сезон')
ax.set_title('Победы долгоиграющих команд')

for index, i in temp.iterrows():
    cur = year_data[year_data.team==i.name]
    plt.plot(cur.season,cur.wins/(cur.played*2))
labels = list(map(lambda x: x[0]+' '+str(x[1]),zip(temp.index,temp.seasons_count)))
ax.legend(labels)
```

```python
temp = state_with_seasons[(state_with_seasons.seasons_count >= 5) & (state_with_seasons.seasons_count <= 6)]
fig, ax = plt.subplots(figsize=(12,6))
ax.set_ylabel('% побед')
ax.set_xlabel('сезон')
ax.set_title('Победы среднеграющих команд')

for index, i in temp.iterrows():
    cur = year_data[year_data.team==i.name]
    plt.plot(cur.season,cur.wins/(cur.played*2))
labels = list(map(lambda x: x[0]+' '+str(x[1]),zip(temp.index,temp.seasons_count)))
ax.legend(labels)
```

```python
temp = state_with_seasons[state_with_seasons.seasons_count < 4]
fig, ax = plt.subplots(figsize=(12,6))
ax.set_ylabel('% побед')
ax.set_xlabel('сезон')
ax.set_title('Победы малоиграющих команд')

for index, i in temp.iterrows():
    cur = year_data[year_data.team==i.name]
    if len(cur) == 1:
        plt.plot(cur.season,cur.wins/(cur.played*2),'r+')
    else:
        plt.plot(cur.season,cur.wins/(cur.played*2))
labels = list(map(lambda x: x[0]+' '+str(x[1]),zip(temp.index,temp.seasons_count)))
ax.legend(labels)
```

Даже визуально можно понять, что средний % побед команд отличается: 0.5-0.6 для долгоиграющих, ~0.3 для средне и ~0.25 для малоиграющих.
